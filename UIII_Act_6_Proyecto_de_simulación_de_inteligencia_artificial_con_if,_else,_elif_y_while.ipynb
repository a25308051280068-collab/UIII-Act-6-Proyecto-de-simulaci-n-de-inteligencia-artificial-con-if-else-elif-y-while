#### Inversión en bolsa

```python
# --- SIMULADOR DE INVERSIÓN AUTÓNOMO (IA AUTOMÁTICA) ---

# Inicialización de variables de estado
saldo = 1000.0       # Dinero inicial
precio_accion = 50.0 # Precio de inicio de la acción
acciones = 0         # Cantidad de acciones compradas
dia = 0              # Contador para el ciclo
limite_dias = 10     # El programa correrá solo por 10 "días"

print("Iniciando simulación automática de inversión...")

# Ciclo principal (while) que sustituye la interacción del usuario [1, 3]
while dia < limite_dias:
    dia = dia + 1
    print(f"\n--- DÍA {dia} ---")

    # 1. Simulación de movimiento de mercado (Lógica automática)
    # Usamos el operador aritmético módulo (%) para variar el precio según el día [4]
    if dia % 2 == 0:
        precio_accion = precio_accion - 5.5  # El precio baja en días pares
    else:
        precio_accion = precio_accion + 8.2  # El precio sube en días impares

    print(f"Precio actual de la acción: ${precio_accion}")

    # 2. Cerebro de la IA (Toma de decisiones automática con if-elif-else) [2]
    # La IA decide comprar si el precio es bajo (Relacional <)
    if precio_accion < 45.0:
        if saldo >= precio_accion:
            # Operaciones aritméticas para ejecutar la compra [4]
            acciones = acciones + 1
            saldo = saldo - precio_accion
            print("IA DECIDE: Precio bajo detectado. COMPRA automática realizada.")
        else:
            print("IA MENSAJE: Quisiera comprar, pero no hay saldo suficiente.")

    # La IA decide vender si el precio es alto (Relacional >)
    elif precio_accion > 55.0:
        if acciones > 0:
            # Operaciones para ejecutar la venta y ganar dinero
            ganancia = acciones * precio_accion
            saldo = saldo + ganancia
            print(f"IA DECIDE: Precio alto detectado. VENTA de {acciones} acciones realizada.")
            acciones = 0
        else:
            print("IA MENSAJE: El precio es bueno para vender, pero no tengo acciones.")

    # Si el precio es medio, la IA no hace nada (Hold) [2]
    else:
        print("IA DECIDE: Precio estable. Manteniendo posición actual (Hold).")

    # Mostrar estado financiero del día
    print(f"Saldo en cuenta: ${saldo:.2f} | Acciones poseídas: {acciones}")

# 3. Resultado final al terminar el ciclo [5]
patrimonio_final = saldo + (acciones * precio_accion)
print("\n==========================================")
print(f"SIMULACIÓN FINALIZADA EN EL DÍA {dia}")
print(f"Patrimonio total obtenido: ${patrimonio_final:.2f}")

if patrimonio_final > 1000.0:
    print("RESULTADO: La IA generó ganancias.")
else:
    print("RESULTADO: La IA generó pérdidas o se mantuvo igual.")
```
